# Quiz เฉลย

```python
len(frequentItemsets[-1][0])
```

```python
two_itemset = []
three_itemset = []
for each in frequentItemsets:
  if len(each[0]) == 2:
    two_itemset.append(each)
  if len (each[0]) == 3:
    three_itemset.append(each)

print(len(two_itemset))
print(len(three_itemset))
```
